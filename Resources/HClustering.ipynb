```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering
import hvplot.pandas
```

```python
iris_df = pd.read_csv('new_iris.csv')
iris_df.head()
```

```python
pca = PCA(n_components=2)
```

```python
iris_scaled = StandardScaler().fit_transform(iris_df)
iris_scaled
```

```python
iris_scaled_pca = pca.fit_transform(iris_scaled)
iris_scaled_pca
```

```python
import plotly.figure_factory as ff
```

```python
# Create the dendrogram
fig = ff.create_dendrogram(iris_scaled_pca, color_threshold=0)
fig.update_layout(width=800, height=500)
fig.show()
```

```python
agg = AgglomerativeClustering(n_clusters=3)
```

```python
model = agg.fit(iris_scaled_pca)
```

```python
model.labels_
```

```python
iris_scaled_pca['class'] = model.labels_
```
